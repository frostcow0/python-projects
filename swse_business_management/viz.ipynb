```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
con = sqlite3.connect("swsebm.db")
```

```python
query = "SELECT * FROM transactions INNER JOIN sessions on transactions.session_id = sessions.session_id"
transactions = pd.read_sql(query, con)
transactions.head()
```

```python
plt.figure(figsize=(14,6))

plt.title("Inventory")

sns.barplot(x=transactions["item"], y=transactions["quantity"])

plt.xlabel("Item")
plt.ylabel("Quantity")

plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.title("Transactions")
plt.xlabel("Dunno")
plt.ylabel("Date")

sns.lineplot(x=transactions["session_date"], y=transactions["quantity"], hue=transactions["item"])

plt.show()
```
