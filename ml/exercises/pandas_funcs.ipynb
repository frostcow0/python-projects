## df.corr()
Creates a cross table showing each column's correlation with each other column.

```python
data.corr()
```

## pd.plotting.scatter_matrix(df)
Also known as a pairs plot, a scatter matrix plots each numerical variable against each other numerical variable.

```python
pd.plotting.scatter_matrix(data, alpha=0.2)
```

# df.hist()

```python
data.hist()
```
