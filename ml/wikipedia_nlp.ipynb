```python
import pandas as pd
```

```python
filename = "text_data/00c2bfc7-57db-496e-9d5c-d62f8d8119e3.json/00c2bfc7-57db-496e-9d5c-d62f8d8119e3.json"
```

```python
df = pd.read_json(filename)
df.describe(include='all')
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
train_text = df['text'].to_list()

tfidf_vect = TfidfVectorizer()

transformed_vect = tfidf_vect.fit_transform(train_text)
```

```python
feat_tfidf = list(zip(
    tfidf_vect.get_feature_names(),
    transformed_vect,
))
feat_tfidf[:5]
```

```python
feat_idf = list(zip(
    tfidf_vect.get_feature_names(),
    tfidf_vect.idf_
))
feat_idf
```

```python
results = pd.DataFrame(feat_idf, columns=['Feature', 'IDF'])
```
