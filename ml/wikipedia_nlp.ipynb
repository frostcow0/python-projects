```python
import pandas as pd
```

```python
filename = "text_data/00c2bfc7-57db-496e-9d5c-d62f8d8119e3.json/00c2bfc7-57db-496e-9d5c-d62f8d8119e3.json"
```

```python
df = pd.read_json(filename)
df.describe(include='all')
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
test = df.iloc[0,1]
test
```

```python
train_text = df.iloc[:3,1]

tfidf_vect = TfidfVectorizer()

transformed_vect = tfidf_vect.fit_transform(train_text)
```

```python
len(tfidf_vect.get_feature_names())
```

```python
feat_tfidf = []
feat_tfidf.append(list(zip(
    tfidf_vect.get_feature_names(),
    transformed_vect.toarray()[0],
)))
feat_tfidf[0]
```

```python
feat_idf = list(zip(
    tfidf_vect.get_feature_names(),
    tfidf_vect.idf_
))
feat_idf
```

```python
results = pd.DataFrame(feat_tfidf[0], columns=['Feature', 'tf-idf'])
sorted_results = results.sort_values(
    by='tf-idf', ascending=False, ignore_index=True)
sorted_results.head(20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()

transformed_count_vect = count_vect.fit_transform(train_text)

sorted_results['Count'] = transformed_count_vect.toarray()[0]
```

```python
newly_sorted_results = sorted_results.sort_values(
    by='Count', ascending=False, ignore_index=True)
newly_sorted_results.head(20)
```
